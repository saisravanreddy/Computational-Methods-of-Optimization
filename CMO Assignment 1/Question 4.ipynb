```python
import numpy as np
```

```python
Q = np.array([
    [2.3346, 1.1384, 2.5606, 1.4507],
    [1.1384, 0.7860, 1.2743, 0.9531],
    [2.5606, 1.2743, 2.8147, 1.6487],
    [1.4507, 0.9531, 1.6487, 1.8123]
             ])

b = np.array([0.4218,0.9157,0.7922,0.9595]).T

x0 = np.array([-135.1150,-4.5224,130.1168,-5.6879]).T 

def quadratic_grad(x):
    return np.matmul(Q,x)-b
    
```

```python
x_opt = np.matmul(np.linalg.inv(Q), b)
print(b.shape)
print(x0.shape)
print(Q.shape)
print(x_opt.shape)
print(x_opt)
```

```python
def gradient_descent(stepsize, epsilon, x):
    no_of_iterations = 0
    while(1):
        delta_x = (-1.)*quadratic_grad(x)
        if np.linalg.norm(x-x_opt) <= epsilon:
            print ("no of iterations: " + str(no_of_iterations))
            return
        else:
            x = x + (stepsize*delta_x)
            no_of_iterations += 1
            print (x, no_of_iterations)
    print ("does not converge with this step size")
```

```python
gradient_descent(0.2,0.0001,x0)
gradient_descent(0.07,0.0001,x0)
gradient_descent(0.0728,0.0001,x0)
gradient_descent(0.2185,0.0001,x0)
```

```python
alpha_opt = -1*(x_opt[0]-x0[0])/(quadratic_grad(x0)[0])
print(alpha_opt)
alpha_opt = -1*(x_opt[1]-x0[1])/(quadratic_grad(x0)[1])
print(alpha_opt)
alpha_opt = -1*(x_opt[2]-x0[2])/(quadratic_grad(x0)[2])
print(alpha_opt)
alpha_opt = -1*(x_opt[3]-x0[3])/(quadratic_grad(x0)[3])
print(alpha_opt)
```
