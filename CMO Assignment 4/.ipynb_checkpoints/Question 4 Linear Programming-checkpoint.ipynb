```python
# Import required libraries
import sys
import numpy as np
from scipy.optimize import linprog
np.set_printoptions(threshold=sys.maxsize)
```

# Problem 4 a)

```python
# Set the inequality constraints matrix
# Note: the inequality constraints must be in the form of <=
#A = np.array([[-1, -1, -1], [-1, 2, 0], [0, 0, -1], [-1, 0, 0], [0, -1, 0], [0, 0, -1]])
A = np.zeros(shape=(40, 12))

for i in range(40):
    A[i,i//4]=-1
    if (i//2)%2==0:
        if (i%2==0):
            A[i,10]=-1
        else:
            A[i,11]=-1
    else:
        if(i%2==0):
            A[i,10]=1
        else:
            A[i,11]=1
    #A[i,i+12]=-1

print(A)
```

```python
from numpy import genfromtxt
X_old = genfromtxt('a4.csv', delimiter=',')
X = X_old.T 
print(X)

# Set the inequality constraints vector
#b = np.array([-1000, 0, -340, 0, 0, 0])
b = np.zeros(shape=(40))
for i in range(40):
    if (i//2)%2==0:
        if (i%2==0):
            b[i]=-X[i//4,0]
        else:
            b[i]=-X[i//4,1]
    else:
        if(i%2==0):
            b[i]=X[i//4,0]
        else:
            b[i]=X[i//4,1]

print(b)


# Set the coefficients of the linear objective function vector
#c = np.array([10, 15, 25])
c = np.zeros(shape=(12))
for i in range(10):
    c[i]=1

print(c)
    
# Solve linear programming problem
res = linprog(c, A_ub=A, b_ub=b)

# Print results
print('Optimal value:', round(res.fun, ndigits=2),
      '\nx values:', res.x,
      '\nNumber of iterations performed:', res.nit,
      '\nStatus:', res.message)
```

# Problem 4 c)

```python
# Set the inequality constraints matrix
# Note: the inequality constraints must be in the form of <=
#A = np.array([[-1, -1, -1], [-1, 2, 0], [0, 0, -1], [-1, 0, 0], [0, -1, 0], [0, 0, -1]])
A = np.zeros(shape=(40, 22))

for i in range(40):
    if (i//2)%2==0:
        if i%2==0:
            A[i,0]=-1
        else:
            A[i,0]=1
    else:
        if i%2==0:
            A[i,1]=-1
        else:
            A[i,1]=1    
    A[i,2+(i//2)]=-1

print(A)
```

```python
from numpy import genfromtxt
X_old = genfromtxt('a4.csv', delimiter=',')
X = X_old.T 
print(X)

# Set the inequality constraints vector
#b = np.array([-1000, 0, -340, 0, 0, 0])
b = np.zeros(shape=(40))
for i in range(40):
    if (i//2)%2==0:
        if i%2==0:
            b[i] = X[i//4,0]
        else:
            b[i] = -X[i//4,0]        
    else:
        if i%2==0:
            b[i] = X[i//4,1]
        else:
            b[i] = -X[i//4,1]

print(b)



# Set the coefficients of the linear objective function vector
#c = np.array([10, 15, 25])
c = np.zeros(shape=(22))
for i in range(2,22):
    c[i]=1

print(c)

# Solve linear programming problem
res = linprog(c, A_ub=A, b_ub=b)

# Print results
print('Optimal value:', round(res.fun, ndigits=2),
      '\nx values:', res.x,
      '\nNumber of iterations performed:', res.nit,
      '\nStatus:', res.message)
```
