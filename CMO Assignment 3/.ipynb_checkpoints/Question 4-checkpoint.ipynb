```python
def newton_point(g,B):

    cho_info = scipy.linalg.cho_factor(B)
    newton_point = -scipy.linalg.cho_solve(cho_info, g)
    return newton_point  ## Title


def f(x):
    A1 = np.matmul(np.matmul(np.transpose(x),A),x)
    print(A1)
    A2 = np.dot(b,X)
    print(A2)
    return np.matmul(np.matmul(np.transpose(x),A),x)+np.dot(b,x)
def  g(x):
    return 2.*np.dot(A,x)+b
```

```python
def minimize_bfgs(f, x0, jac=None,gtol=1e-5,disp=False):
    # The first step, given the initial point x0, Ho, calculate go
    gfk = g(x0)
    k = 0
    N = len(x0)
    I = np.eye(N, dtype=int)
    Bk = I
    # Initialize the Hessian matrix
    xk = x0
    # The initial direction is the fastest descending direction

    gnorm = np.amax(np.abs(gfk))
    while (gnorm > gtol):
        # Second step: Find the descending direction
        # pk = -numpy.dot(Hk, gfk) 
        pk=newton_point(gfk,Bk)
        # Third step: Here we use precise line search to determine the step size, and update xk+1
        alpha_k = -np.dot(pk.T, g(xk))/np.dot(pk.T,np.dot(A,pk))
        xkp1 = xk + alpha_k * pk
        # Fourth step: Correction of Bk
        sk = xkp1 - xk
        xk = xkp1
        gfkp1 = g(xkp1)
        yk = gfkp1 - gfk
        gfk = gfkp1
        k += 1
        gnorm = np.amax(np.abs(gfk))

        if (gnorm <= gtol):
            break
        rhok = 1.0 / (np.dot(yk, sk))
        A1= I - yk[:, None] *sk[None, :] * rhok
        A2 = I - sk[:, None] * yk[None, :] * rhok

        Bk = np.dot(np.dot(A1, Bk),A2 ) + rhok * sk[:,None]*sk[None,:]




        print("Current function value xk:[%f %f]"% (xk[0],xk[1]))
        print("Current function value: %f" % f(xk))
        print("BK",Bk)

```
