# Question 1 c)

```python
def rosen_2d_grad(X):
    x = X[0]
    y = X[1]
    a = 1. - x
    b = y - x*x
    return np.array([-2*a - 400*b*x, 200*b])

def rosen_2d_Hess(X):
    x = X[0]
    y = X[1]
    h11=2-400*y + 1200*x*x
    h12=-400*x
    h22=200
    return np.array([[h11,h12],[h12,h22]])


def newton_iteration_method(x):
    no_of_iterations = 0
    while(no_of_iterations<2):
        delta_x = (-1.)*rosen_2d_grad(x)
        x = x + np.matmul(np.linalg.inv(rosen_2d_Hess(x)),delta_x)
        no_of_iterations += 1
        print ("X: ["+ str(x[0])+ "," + str(x[1]) + "]")

newton_iteration_method(np.array([0,0]))
```

# Question 1)d

```python
newton_iteration_method(np.array([100,100]))
```

# Yes it converges to [1,1]

# Question 1)e

```python
def gradient_descent(stepsize, epsilon, x):
    no_of_iterations = 0
    while(no_of_iterations<20000):
        delta_x = (-1.)*rosen_2d_grad(x)
        if np.linalg.norm(delta_x) <= epsilon:
            print ("no of iterations: " + str(no_of_iterations))
            return
        else:
            x = x + (stepsize*delta_x)
            no_of_iterations += 1
            print ("X: ["+ str(x[0])+ "," + str(x[1]) + "]")
    print ("does not converge with this step size")
    
gradient_descent(stepsize=0.05, epsilon=0.001, x=np.array([100,100]))
```
